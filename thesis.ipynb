## StopWords
Stopwords are the most common words in any natural language. For the purpose of analyzing text data and building NLP models, these stopwords might not add much value to the meaning of the document. They certainly do not add much value to the sentiments expressed in the text. Generally, the most common words used in a text are ‚Äúthe‚Äù, ‚Äúis‚Äù, ‚Äúin‚Äù, ‚Äúfor‚Äù, ‚Äúwhere‚Äù, ‚Äúwhen‚Äù, ‚Äúto‚Äù, ‚Äúat‚Äù etc. These are examples of stopwords

```python
import nltk
# from nltk.corpus import stopwords
```

```python
stopwords_file = open("stopwords.txt", "r")
try:
    content = stopwords_file.read()
    stopwords = content.split(",")
finally:
    stopwords_file.close()
```

```python
stopwords=[i.replace('"',"").strip() for i in stopwords]
print(len(stopwords))
```

```python
# To properly clean the whole dataset, we create a cleaner function
def dataset_cleaner(text):
    new_text = re.sub("@[A-Za-z0-9]+", "", text) #removing the @ tags
    new_text = re.sub("[\n]", " ", new_text) # removing newline tags
    new_text = re.sub(r"http\S+", "", new_text) # removing hyperlinks
    new_text = contractions.fix(new_text) # fixing contractions
    new_text = re.sub("#", "", new_text) #removing hashtags
    new_text = re.sub(r"[^a-zA-Z]", " ", new_text) # removing all non-alphabets
    new_text = new_text.lower().strip() # converting text to lower case
    
    cleaned_text = ''
    # lemmatizing tokens and removing stopwords
    for token in new_text.split():
        if token not in  stopwords:
            cleaned_text = cleaned_text + lemmatizer.lemmatize(token) + ' '
    
    return cleaned_text
```

```python
merged_df['Clean Data'] = merged_df['Data'].apply(lambda x: dataset_cleaner(x))
```

### Using Lexicon-Based VADER tool

```python
from afinn import Afinn

afn = Afinn()
```

```python
afinn_df = pd.DataFrame()
afinn_df['Data'] = merged_df['Clean Data']
afinn_df
```

```python
afinn_df['afinn_score'] = afinn_df['Data'].apply(lambda x: afn.score(x))
```

```python
afinn_df
```

```python
# converting afinn score to sentiment classes 1, 0, -1
def afinn_conv(x):
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
```

```python
afinn_df['class'] = afinn_df['afinn_score'].apply(lambda x: afinn_conv(x))
```

```python
afinn_df 
```

```python
print(classification_report(merged_df['Labell'], afinn_df['class']))
```

```python
afinn_accuracy = accuracy_score(merged_df['Labell'], afinn_df['class'])
print(f"(Accuracy: {afinn_accuracy})")
afinn_precision = precision_score(merged_df['Labell'], afinn_df['class'], average='weighted')
print(f"(Precision: {afinn_precision})")
afinn_recall = recall_score(merged_df['Labell'], afinn_df['class'], average='weighted')
print(f"(Recall: {afinn_recall})")
afinn_f1 = f1_score(merged_df['Labell'], afinn_df['class'], average='weighted')
print(f"(F1 Score: {afinn_f1})")
```
